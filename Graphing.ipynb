```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 10 #Number of users

mu = 2 #Service rate 2

lambdas = np.linspace(0, 20, 100) #We Generate a range of lambda (arrival rate) values
```

```python
# Calculate mean response times for the three different systems at each lambda value
rs1 = [l / (m * mu) for l in lambdas]
mrts1 = [(1 / mu) / (1 - rho) if rho < 1 else 1.5 for rho in rs1]  

rs2 = [l / (m * mu) for l in lambdas]
mrts2 = [(1 / (m * mu - l)) if 1 / (m * mu - l) < 1.5 else 1.5 for l in lambdas] 

mrts3 = [(1 / mu) / (1 - (l / (m * mu))) * (1 / (1 + l / ((m - 1) * mu))) if 1 / mu / (1 - (l / (m * mu))) * (1 / (1 + l / ((m - 1) * mu))) < 1.5 else 1.5 for l in lambdas]
```

```python
plt.plot(lambdas, mrts1, label='Upgrade 1')
plt.plot(lambdas, mrts2, label='Upgrade 2')
plt.plot(lambdas, mrts3, label='Upgrade 3')

plt.xlabel('Arrival Rate ()')
plt.ylabel('Mean Response Time')
plt.title('Mean Response Times for Three Different Systems')
plt.xlim(1, 20)
plt.ylim(0, 1.5)

plt.legend()
plt.show()
```
